```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def linear(x):
    return x
```

```python
def step(x):
    return 0 if x < 0 else 1
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def tanh(x):
    return np.tanh(x)
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
def Lrelu(x):
    return np.maximum(0.1*x, x)
```

```python
def elu(x,a):
    return np.maximum(x, a*(np.exp(x) - 1 ))
```

```python
def softplus(x):
    return (1+np.log(np.exp(x)))
```
